### Question

You and a friend take turns rolling a fair six-sided die, and the first person to roll a 
6 wins. Note that the person who rolls first doesn't have to win on the first roll. He or she merely has to win eventually.

What's the probability that the person who makes the first roll wins the game?

#### Answer

- P(1) = prob of the first person roll a 6 and win
- P(1_) = prob of not rolling a 6, the game goes on 

- P(2)= prob of the second person roll a 6 and win
- P(2_) = prob of not rolling a 6, the game goes on 

Probability of the first person roll and win ON THE FIRST ROLL:
>P(1win|1st roll) = P(1) = 1/6

Probability of the first person roll and win ON THE SECOND ROLL:
>P(1win|2nd roll) = P(1_)*P(2_) * P(1) = (5/6 * 5/6) * 1/6

Probability of the first person roll and win ON THE THIRD ROLL:
>P(1win|3rd roll) = P(1_)*P(2_) *  P(1_)*P(2_) * P(1) = (5/6 * 5/6)^2 * 1/6

... and so on,.

The series is:

Apply the formular, we have:

P(1win) = (1/6) / (1-25/36)

```python
(1/6) / (1-25/36)
```

### Question

Stephen is fishing at Lake Andrews. Let's assume the probability that he catches at least 
1 fish in any given hour is uniform and independent across time. The probability that Stephen catches at least 
1 fish in 
2 hours of fishing is 
64%.

What's the probability that he catches at least 
1 fish in an hour of fishing?



**Answer:**

Chance of catching at least 1 fish in an hour = 1 - chance of catching no fish in 1 hour

chance of catching at least 1 fish in 2 hours = 1 - Chance of catching no fish in 2 hour = 1 - (chance of catching no fish in 1 hour)^2

=> chance of catching no fish in 1 hour = square root (1 - chance of catching at least 1 fish in 2 hours)

```python
# chance of catch no fish in 2 hours:
1-0.64
```

```python
# chance of catch no fish in 1 hour:
sqrt(0.36)
```

```python
# chance of catch at least 1 fish in 1 hour:
1-0.6
```
